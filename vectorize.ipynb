```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

import preprocessing
```

```python
documents = preprocessing.get_clean_tweets(subset=10000)
```

```python
documents[0]
```

```python
documents[5000]
```

```python
cv = CountVectorizer(stop_words='english')

cv_array = cv.fit_transform(documents).toarray()

cv_frame = pd.DataFrame(cv_array, columns = cv.get_feature_names())
```

```python
cv_frame.head()
```

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_array = tfidf.fit_transform(documents).toarray()

tfidf_frame = pd.DataFrame(tfidf_array, columns=tfidf.get_feature_names())
```
